```python
import pandas as pd
import numpy as np
```

```python
# Creating raw dataset
data = {'property_id':[1,2,3,4,5], 
'price':[250000,300000,150000,400000,350000],
'size_sqft':[1500,2000,1200,2500,1800],
'date_listed':['2023-01-15','2023-02-20','2023-03-10','2023-04-05','2023-05-01'],
'year_built':[1990,2005,1980,2010,2000],
'property_type':['Apartment','Villa','Apartment','Villa','Independent House'],
'location':['Vizag','Hyderabad','Vizag','Bangalore','Chennai']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Price per sqft
df['price_per_sqft'] = df['price'] / df['size_sqft']
df['price_per_sqft'] = df['price_per_sqft'].round(2)
df
```

```python
#data transformation
df['date_listed'] = pd.to_datetime(df['date_listed'])
df
```

```python
df['month'] = df['date_listed'].dt.month   # Extracting month from date_listed
df
```

```python
df['listing_year'] = df['date_listed'].dt.year # Extracting year from date_listed
df
```

```python
from sklearn.preprocessing import MinMaxScaler #normalization
scaler = MinMaxScaler()
df['normalized_price'] = scaler.fit_transform(df[['price']])
df
```

```python
from sklearn.preprocessing import StandardScaler #standardization
scaler = StandardScaler()
df['standard_size'] = scaler.fit_transform(df[['size_sqft']])
df
```

```python
df = pd.get_dummies(df, columns=['property_type']) # One-hot encoding for property_type
df
```

```python
from sklearn.preprocessing import LabelEncoder # Label encoding for location
le = LabelEncoder()
df['location_encoded'] = le.fit_transform(df['location'])
df
```

```python
Q1 = df['price'].quantile(0.25) #outlier detection using IQR
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[ (df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR))
]
outliers
```
